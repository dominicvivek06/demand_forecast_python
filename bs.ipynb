```python
import requests
import lxml
import bs4
import pandas as pd
```

```python
web = requests.get('https://docs.aws.amazon.com/translate/latest/dg/what-is.html')
```

```python
soup = bs4.BeautifulSoup(web.text,'lxml')
```

```python
table = soup.find('table', attrs={'id': 'w117aab5b5b7'})
table
```

```python
columns = ['language','code']
df = pd.DataFrame(columns=columns)
trs = table.find_all('tr')
for tr in trs:
    tds = tr.find_all('td')
    row = [td.text.replace('\n','') for td in tds]
    if len(row) != 0:
        #print(row,len(row))
        df = df.append(pd.Series(row, index=columns), ignore_index=True)
```
